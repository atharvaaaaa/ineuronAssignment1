# 10. How can you tell the difference between break and continue?
Ans: Break and Continue statements are used in loops

Break: We use Break in the loop to stop the loop when the given conditions are met

ex: 

a = 1

for(a<=5):

    print(a)
    
    
    if(a==3):
    
        break
    a=a+1
    
        
Output: 

1
2
3

Continue: 
We use Break in the loop to end the iteration when given conditions are met and continue to next iteration

ex:

a = 1

for(a<=5):

    print(a)
    
    
    if(a==3):
    
        Continue
    a=a+1
        
Output:

1
2
4
5
